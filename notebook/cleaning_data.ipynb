```python
import pandas as pd
```

```python
PATH = "../data/dataset/continuous dataset.csv"
```

```python
df = pd.read_csv(PATH)
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Convert datetime
df['datetime'] = pd.to_datetime(df["datetime"])
```

```python
df['hour'] = df['datetime'].dt.hour
df['day'] = df['datetime'].dt.day
df['month'] = df['datetime'].dt.month
df['day_of_week'] = df['datetime'].dt.weekday
df['is_weekend'] = df['day_of_week'].apply(lambda x: 1 if x >= 5 else 0)
```

```python
df.head()
```

```python
PATH_SAVE = "../data/dataset"
```

```python
df.to_csv(f'{PATH_SAVE}/cleaning_dataset.csv', index=False)
```
